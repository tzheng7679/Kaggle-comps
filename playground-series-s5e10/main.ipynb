## Building Test Network

```python
import torch
from torch import nn
from torch.utils.data import DataLoader
```

```python
class NeuralNetwork(nn.Module):
    def __init__(self, d_in):
        super().__init__()
        
        self.layers = nn.Sequential(
            nn.Linear(d_in, 64),
            nn.ReLU(),
            nn.Linear(64, 1)
        )
    
    def forward(self, x):
        return self.layers(x)

```

Grab data that is numeric for test (deal with categorical features later)

```python
numeric = train.select_dtypes([int, float])
```

```python
device = torch.accelerator.current_accelerator() if torch.cuda.is_available() else "cpu"
model = NeuralNetwork(len(numeric.columns)).to(device)
model.double()
```

```python
model(torch.from_numpy(numeric.to_numpy()))
```
